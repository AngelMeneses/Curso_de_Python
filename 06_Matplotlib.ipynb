## Matplotlib: Análisis y Visualización de Datos


[Matplotlib](http://matplotlib.org/) es una excelente biblioteca de gráficos 2D y 3D; los aspectos de las figuras pueden ser contraladas con programación. Algunas características son:

- Se puede usar $\LaTeX$ para etiquetas y textos. 
- Se puede cambiar el tamaño de la figura y los DPI. 
- Salida de alta calidad en muchos formatos: PNG, PDF, SVG, EPS y PGF.

`Matplotlib.pyplot` es un submódulo donde funciona como MATLAB.

<div class="alert alert-info">
Además de <a href="http://matplotlib.org/"> Matplotlib </a> existen otras utilizadas: <a href="https://pypi.org/project/vpython/">VPython</a> y <a href="https://pypi.org/project/mayavi/">mayavi</a>.

```python
import numpy as np
import matplotlib
# Comúnmente se le coloca el alias de plt
from matplotlib import pyplot as plt
from numpy import random
%who
```

```python
# Se crean las dos variables de forma aleatoria
x = random.rand(4)
y = random.rand(4)
```

```python
# plot():
plt.plot(x, y)
```

```python
## Como si fuera Matlab
plt.figure()
plt.plot(x, y, 'b')
plt.xlabel('eje x')
plt.ylabel('eje y')
plt.title('título $\mu$')
plt.show()
```

```python
plt.plot(x, y, 'b--', label="$x$", alpha=1)
plt.plot(x-y, y, label="$x-y$", alpha=1)
plt.xlabel("eje y", fontsize=15)
plt.ylabel("eje x", fontsize=15)
plt.title("Dos en una")
plt.legend(loc=1)
```

```python
# Se importa todas las funciones para no usar plt.*
from matplotlib.pyplot import *
```

```python
x_1 = random.rand(4)
y_1 = random.rand(4)
```

```python
subplot(1,2,1)
plot(x, y, 'b-o')
subplot(1,2,2)
plot(x_1, y_1, 'r*-');
```

```python
subplot(2,2,1)
plot(x, y, 'b-o')
subplot(2,2,2)
plot(x_1, y_1, 'g*-');
subplot(2,2,3)
plot(y, x, 'b--')
subplot(2,2,4)
plot(y_1, x_1, 'g*--');
```

## Matplotlib orientado a objetos

Principalmente la programación orientada a objetos es tener objetos para aplicar funciones, operaciones o acciones. En matplotlib, se puede definir una figura global creando la variable`fig`, y con esta creamos los `axes` usando `add_axes`.

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 1, 1]) # left, bottom, width, height (range 0 to 1)

axes.plot(x, y)
axes.set_xlabel('eje x')
axes.set_ylabel('eje y')
axes.set_title('título')
```

Ahora tenemos un control de los ejes del gráfico, y podemos añadir más de un eje a la figura:

```python
fig = plt.figure()

# Se definen las dimensiones de los gráficos
axes1 = fig.add_axes([0.5, 0.1, 1, 1]) 
axes2 = fig.add_axes([0.6, 0.6, 0.4, 0.4]) 

# Gráfico principal
axes1.plot(x, y, 'b')
axes1.set_xlabel('eje x')
axes1.set_ylabel('eje y')
axes1.set_title('Gráfico principal')

axes2.plot(y, x, 'r--')
axes2.set_xlabel('eje y')
axes2.set_ylabel('eje x')
axes2.set_title('Gráfico pequeño');
```

```python
# una sola gráfica.
## Se define fig y axes al mismo tiempo
fig, axes = plt.subplots()

axes.plot(x, y, 'b')
axes.set_xlabel('eje x')
axes.set_ylabel('eje y')
axes.set_title('Subplot 1');
```

```python
# varias gráficas
fig, axes = plt.subplots(ncols=2)

axes[0].plot(x, y, 'b')
axes[1].plot(x-y, y, 'g')
```

```python
# Haciendo uso del for para crear varias columnas y filas
fig, axes = plt.subplots(nrows=1, ncols=2)

for ax in axes:
    ax.plot(x,y, 'm')
    ax.set_xlabel('eje x')
    ax.set_ylabel('eje y')
    ax.set_title('Mismo título para las dos subfiguras')
```

- Genera las dos subfiguras, aunque los ejes están uno sobre el otro, para solucionar esto se usa la función `fig.tight_layout`

```python
# fig.tight_layout()
fig, axes = plt.subplots(nrows=1, ncols=2)

for ax in axes:
    ax.plot(x,y, 'm')
    ax.set_xlabel('eje x')
    ax.set_ylabel('eje y')
    ax.set_title('Es el mismo título para las dos gráficas')
fig.tight_layout()
```

- ¿Y los títulos?

```python
# figsize: recibe una tupla de ancho y altura, respectivamente.
fig, axes = plt.subplots(nrows=1, ncols=2, figsize = (10,5))

for ax in axes:
    ax.plot(x,y, 'm')
    ax.set_xlabel('eje x')
    ax.set_ylabel('eje y')
    ax.set_title('Es el mismo título para las dos gráficas')
fig.tight_layout()
```
