```python
# https://www.codechef.com/problems/MARBLES
# 925
# Permutaciones y Combinaciones, Particiones
# Matem√°ticas, Combinatoria

def factorial(x):
    fact = 1
    for i in range(x):
        fact+= fact*i
    return fact

for ts in range(int(input())):
    (N, K) = map(int, input().split(' '))
    comb = factorial(N-1)/(factorial(K-1)*factorial(N-K))
    print(comb)
```

```python
# https://www.codechef.com/problems/DARLIG
# 994
# Case work, Special
T = int(input())
l =[1, 1, 1, 0]

for ts in range(T):
    (N, K) = map(int, input().split(' '))
    list_2 = []
    for i in l:
        list_2.append(i==K)
        
    init = []
    for i in range(len(list_2)):
        if list_2[i] == True:
            init.append(i)
    
    result = []
    for i in init:
        if i+N <=3:
            result.append(i+N)
        else:
            if i+N%4 <= 3:
                result.append(i+N%4)
            else:
                result.append(i+N%4-4)
    result_1 = []
    for i in result:
        if i <3:
            result_1.append(1)
        else:
            result_1.append(0)
            
    result_2 = [i for i in set(result_1)]
    
    if len(result_2) == 1:
        if result_2[0] == 1:
            print('On')
        else:
            print('Off')
    else: 
        print('Ambiguous')
```

```python
# https://www.codechef.com/problems/KEPLERSLAW
# 992
# cakewalk, jan221, srikkanth_adm
T = int(input())
for ts in range(T):
    (T1, T2, R1, R2) = map(int, input().split(' '))
    if (T1**2/R1**3) == (T2**2/R2**3):
        print('Yes')
    else:
        print('No')
```

```python
# https://www.codechef.com/problems/CAN_REACH
#  990
# Conditional Statements, Cartesian Coordinate System, 
# Divisibility, ad-hoc
# Basic Programming Concepts, Mathematics, 
# Modular Arithmetic, Algorithms, Constructive
T = int(input())
x,y = 0,0
for t in range(T):
    (a, b, K) = map(int, input().split(' '))
    List = []
    if x < a:
        i = x + K
        while True:
            if i == a:
                List.append(True)
                break
            elif i <= a:
                i += K
            else:
                List.append(False)
                break
    if x > a:
        i = x - K        
        while True:
            if i == a:
                List.append(True)
                break
            elif i >= a:
                i -= K
            else:
                List.append(False)
                break
    if y < b:
        i = y + K        
        while True:
            if i == b:
                List.append(True)
                break
            elif i <= b:
                i += K
            else:
                List.append(False)
                break
                
    if y > b:                
        i = y - K        
        while True:
            if i == b:
                List.append(True)
                break
            elif i >= b:
                i -= K
            else:
                List.append(False)
                break
    
    if (List[0] == True) and (List[1]==True):
        print('YES')
    else:
        print('NO')
```

```python
# https://www.codechef.com/problems/LOOP
# 838
# Basic Maths, Implementation
# Mathematics, Algorithms
T = int(input())
for ts in range(T):
    (A,B,M) = map (int, input().split(' '))
    if (A < B):
        a = B - A
        b = M + A - B
        if (a < b):
            print(a)
        else:
            print(b)
    else:
        a = A - B
        b = M - A + B
        if (a < b):
            print(a)
        else:
            print(b)
```
